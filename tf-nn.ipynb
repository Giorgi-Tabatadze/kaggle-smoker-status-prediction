The Goal of this notebook is to predict the smoker status of an individual through
the different parameters that are given. 
This is a classical binary classficiation task.

```python
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras.layers import Dropout
```

```python
df = pd.read_csv("./playground-series/train.csv")
test_data = pd.read_csv('./playground-series/test.csv')
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['hearing(left)', 'hearing(right)', 'Urine protein', 'dental caries'], dtype=int)
test_data = pd.get_dummies(test_data, columns=['hearing(left)', 'hearing(right)', 'Urine protein', 'dental caries'], dtype=int)
```

```python
train_y = df.smoking
train_x = df.drop(columns=["smoking", "id"])
train_x, dev_x, train_y, dev_y = train_test_split(train_x, train_y, test_size=0.05, random_state=42)
#scaler = StandardScaler()
#train_x_scaled = scaler.fit_transform(train_x)
#dev_x_scaled = scaler.transform(dev_x)
#test_x_scaled = scaler.transform(test_data.drop(columns='id'))
```

```python
input_layer = keras.Input(shape=(train_x.shape[1],))

x = keras.layers.Dense(128, kernel_initializer='he_normal')(input_layer)
x = keras.layers.BatchNormalization()(x)
x = keras.layers.Activation('relu')(x)
x = Dropout(0.1)(x) 


x = keras.layers.Dense(64, kernel_initializer='he_normal')(x)
x = keras.layers.BatchNormalization()(x)
x = keras.layers.Activation('relu')(x)
x = Dropout(0.1)(x) 

output_layer =keras.layers.Dense(1, activation='sigmoid')(x)

model = keras.Model(inputs=input_layer, outputs=output_layer)
```

```python
model.compile(optimizer=keras.optimizers.legacy.Adam(0.1),
              loss='binary_crossentropy',
              metrics=['accuracy', tf.keras.metrics.AUC(name='auc')])
```

```python
model.fit(train_x, train_y, epochs=30, batch_size=32, validation_data=(dev_x, dev_y))
```
